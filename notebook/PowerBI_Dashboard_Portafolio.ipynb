# 📊 Power BI - Dashboards Interactivos
Este notebook presenta capturas y descripción de dashboards construidos con Power BI para análisis mensual de folios por ingeniero y zona.

## 📌 Dashboard General
🧾 Descripción
Este dashboard presenta una visión ejecutiva del desempeño mensual de los ingenieros encargados de la resolución de folios (tickets), desglosado por tipo de folio (Incidencias, Servicio, Reparación) y por ingeniero, permitiendo una rápida interpretación de los volúmenes y distribución del trabajo en la organización.

📌 Características principales:

-Total de folios acumulados con tarjeta de resumen.

-Gráfico de columnas: Visualización de folios acumulados por cada ingeniero.

-Gráfico de barras horizontales: Distribución de folios por tipo (Incidencias, Servicio, Reparación).

-Tabla resumen dinámica que agrupa:

-Nombre del ingeniero

-Zona a la que pertenece

-Cantidad de folios por tipo

🧠 Análisis realizado:

-Comparativo de carga laboral entre ingenieros.

-Identificación de zonas o ingenieros con mayor volumen de solicitudes.

-Distribución porcentual del tipo de servicios solicitados.

🌱 Posibles datos a agregar en el futuro:

-Porcentaje de cumplimiento por tipo de folio.

-Promedio de resolución por tipo de folio.

-Clasificación por prioridad de folio.

-Estado de los folios (resueltos/no resueltos) por tipo.

-Fecha de creación y fecha de cierre para análisis temporal.

## 👤 Dashboard por Ingeniero
🧾 Descripción

Este dashboard está diseñado para mostrar un análisis individualizado de desempeño por ingeniero, facilitando una evaluación rápida de su eficacia, puntualidad y cumplimiento de estándares en la resolución de folios.

📌 Características principales:

-Ingeniero seleccionado dinámicamente mediante filtro.

-KPI’s clave:

-% Resolución

-% Puntualidad

-Tarjetas de cantidad de folios acumulados y resueltos/no resueltos.

-Gráfico de barras: Cantidad por tipo de folio (Incidencia, Servicio, Reparación).

-Indicadores tipo gauge para mostrar el cumplimiento visual.

-Gráfico de dona para analizar la puntualidad:

-Folios resueltos en tiempo

-Al límite de tiempo

-Excedidos

🧠 Análisis realizado:

-Desempeño individual en resolución de folios.

-Evaluación visual de la puntualidad del ingeniero.

-Balance de carga y cumplimiento del SLA (Service Level Agreement).

🌱 Posibles datos a agregar en el futuro:

-Comparativa histórica de desempeño del ingeniero (últimos 3-6 meses).

-Calificaciones de satisfacción del cliente.

-Detalle de folios críticos o urgentes.

-Tipo de equipo o área técnica atendida.

-Rendimiento frente al promedio del equipo o zona.
